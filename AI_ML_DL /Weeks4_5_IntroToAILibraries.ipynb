## **Understanding Tensors and Basic Operations**

Task: Dive deeper into tensor operations using both TensorFlow and PyTorch. Practice creating, reshaping, and performing basic mathematical functions with tensors.

### Basic Tensor Operations

#### Creating Tensors in TensorFlow and PyTorch

```python
# TensorFlow
import tensorflow as tf

# Creating a tensor
tensor_tf = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)
print("TensorFlow Tensor:\n", tensor_tf)
```

```python
# PyTorch
import torch

# Creating a tensor
tensor_torch = torch.tensor([[5, 6], [7, 8]], dtype=torch.float32)
print("PyTorch Tensor:\n", tensor_torch)
```

#### Basic Arithmetic Operations

```python
# TensorFlow

# Addition
add_tf = tensor_tf + 2
print("\nTensorFlow Addition:\n", add_tf)

# Subtraction
sub_tf = tensor_tf - 2
print("\nTensorFlow Subtraction:\n", sub_tf)

# Element-wise multiplication
mul_tf = tensor_tf * 2
print("\nTensorFlow Multiplication:\n", mul_tf)
```

```python
# PyTorch

# Addition
add_torch = tensor_torch + 2
print("\nPyTorch Addition:\n", add_torch)

# Subtraction
sub_torch = tensor_torch - 2
print("\nPyTorch Subtraction:\n", sub_torch)

# Element-wise multiplication
mul_torch = tensor_torch * 2
print("\nPyTorch Multiplication:\n", mul_torch)
```

### Reshaping Tensors

```python
# Reshaping the tensor - TensorFlow
reshaped_tf = tf.reshape(tensor_tf, [4, 1])
print("\nReshaped TensorFlow Tensor:\n", reshaped_tf)
```

```python
# Reshaping the tensor - PyTorch
reshaped_torch = tensor_torch.view(4, 1)
print("\nReshaped PyTorch Tensor:\n", reshaped_torch)
```

### Matrix Multiplication

```python
# Matrix multiplication in TensorFlow
result_tf = tf.matmul(tensor_tf, tf.transpose(tensor_tf))
print("\nMatrix Multiplication Result in TensorFlow:\n", result_tf)
```

```python
# Matrix multiplication in PyTorch
result_torch = torch.matmul(tensor_torch, tensor_torch.T)
print("\nMatrix Multiplication Result in PyTorch:\n", result_torch)
```

### Basic Math Operations

```python
# Sum and Mean in TensorFlow
sum_tf = tf.reduce_sum(tensor_tf)
mean_tf = tf.reduce_mean(tensor_tf)
print("\nSum of TensorFlow Tensor:", sum_tf.numpy())
print("Mean of TensorFlow Tensor:", mean_tf.numpy())
```

```python
# Sum and Mean in PyTorch
sum_torch = torch.sum(tensor_torch)
mean_torch = torch.mean(tensor_torch)
print("\nSum of PyTorch Tensor:", sum_torch.item())
print("Mean of PyTorch Tensor:", mean_torch.item())
```

## **Building a Simple Neural Network Model**

Task : Create a simple neural network for basic understanding

### Simple Neural Network in Keras

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Generate simple synthetic data for classification
X_train = np.random.rand(500, 5)  # 500 samples, 5 features each
y_train = np.random.randint(0, 2, 500)

# Build a simple model with one hidden layer
model = Sequential([
    Dense(16, activation='relu', input_shape=(5,)),  # Input layer with 5 features
    Dense(8, activation='relu'),  # Hidden layer with 8 units and ReLU activation
    Dense(1, activation='sigmoid')  # Output layer with 1 unit and sigmoid activation for binary classification
])
# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=5, batch_size=16, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(X_train, y_train)
print("\nSimple Model Accuracy:", accuracy)
```

```python
model.summary()
```

# **Week 5: Building a Fully Connected Neural Network**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import numpy as np

# Generate larger synthetic data for classification
X_train = np.random.rand(50000, 5)  # 50,000 samples, 5 features each
y_train = np.random.randint(0, 2, 50000)  # Binary target (0 or 1)

# Build a more complex model with two hidden layers and dropout
model = Sequential([
    Dense(32, activation='relu', input_shape=(5,)),  # Input layer with 5 features
    Dropout(0.2),  # Dropout layer with 20% rate
    Dense(16, activation='relu'),  # First hidden layer with 16 units
    Dropout(0.2),  # Dropout layer
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Display model summary
model.summary()
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=16, validation_split=0.2)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_train, y_train)
print("\nFully Connected Model Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Generate predictions
y_pred = (model.predict(X_train) > 0.5).astype("int32")

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_train, y_pred))

# Print the confusion matrix
cm = confusion_matrix(y_train, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
from tensorflow.keras.optimizers import Adam

# Compile the model with a custom learning rate
optimizer = Adam(learning_rate=0.001)  # Change the learning rate here
model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=16, validation_split=0.2)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_train, y_train)
print("\nFully Connected Model Accuracy:", accuracy)

# Generate predictions
y_pred = (model.predict(X_train) > 0.5).astype("int32")

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_train, y_pred))

# Print the confusion matrix
cm = confusion_matrix(y_train, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
# Compile the model with the default Learning rate
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model with a different batch size
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)  # Change batch size
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X_train, y_train)
print("\nFully Connected Model Accuracy:", accuracy)

# Generate predictions
y_pred = (model.predict(X_train) > 0.5).astype("int32")

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_train, y_pred))

# Print the confusion matrix
cm = confusion_matrix(y_train, y_pred)
print("\nConfusion Matrix:")
print(cm)
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
history2 = model.fit(X_train, y_train, epochs=10, batch_size=16, validation_split=0.2)
```

```python
# Extracting metrics for batch_size=32
accuracy1 = history.history['accuracy']
val_accuracy1 = history.history['val_accuracy']
loss1 = history.history['loss']
val_loss1 = history.history['val_loss']

# Extracting metrics for batch_size=16
accuracy2 = history2.history['accuracy']
val_accuracy2 = history2.history['val_accuracy']
loss2 = history2.history['loss']
val_loss2 = history2.history['val_loss']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))

# Plot Accuracy
plt.subplot(2, 1, 1)
plt.plot(accuracy1, label='Train Accuracy (batch_size=32)')
plt.plot(val_accuracy1, label='Val Accuracy (batch_size=32)', linestyle='dashed')
plt.plot(accuracy2, label='Train Accuracy (batch_size=16)')
plt.plot(val_accuracy2, label='Val Accuracy (batch_size=16)', linestyle='dashed')
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend()
plt.grid()

# Plot Loss
plt.subplot(2, 1, 2)
plt.plot(loss1, label='Train Loss (batch_size=32)')
plt.plot(val_loss1, label='Val Loss (batch_size=32)', linestyle='dashed')
plt.plot(loss2, label='Train Loss (batch_size=16)')
plt.plot(val_loss2, label='Val Loss (batch_size=16)', linestyle='dashed')
plt.title('Model Loss Comparison')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
